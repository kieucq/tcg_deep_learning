```python
from tensorflow import keras
from tensorflow.keras import layers
print("Ready")
```

```python
import os
import matplotlib.pyplot as plt
from tensorflow.keras.utils import load_img, img_to_array
import cv2
input_dir = "/N/project/pfec_climo/ckieu/data/python/images/"
target_dir = "/N/project/pfec_climo/ckieu/data/python/annotations/trimaps/"
input_img_paths  = sorted([os.path.join(input_dir,fname) for fname in os.listdir(input_dir) if fname.endswith(".jpg")])
target_img_paths = sorted([os.path.join(target_dir,fname) 
                           for fname in os.listdir(target_dir) if fname.endswith(".png") and not fname.startswith(".")])
print(input_img_paths[9])
plt.axis("off")
plt.imshow(load_img(input_img_paths[9]))
plt.figure()
test = cv2.imread(input_img_paths[10])
plt.imshow(test)
plt.show()
```

```python
def display_target(target_array):
    normalized_array = (target_array.astype("uint8")-1)*127
    plt.axis("off")
    plt.imshow(normalized_array[:,:,0])

print(target_img_paths[9])    
img = img_to_array(load_img(target_img_paths[9],color_mode="grayscale"))
display_target(img)
```

```python
import numpy as np
import random

img_size = (128,128) # very important to choose the right img size so conv2d and conv2dtranspose can match
num_imgs = len(input_img_paths)
random.Random(1337).shuffle(input_img_paths)
random.Random(1337).shuffle(target_img_paths)

def path_to_input_image(path):
    img = img_to_array(load_img(path, target_size=img_size))
    return img

def path_to_target_image(path):
    img = img_to_array(load_img(path,target_size=img_size, color_mode="grayscale"))
    img = img.astype("uint8") - 1
    return img

input_shape=(num_imgs,)+img_size+(3,)
print(input_shape)
input_imgs = np.zeros(input_shape,dtype="float32")
targets = np.zeros((num_imgs,)+img_size+(1,),dtype="uint8")
targets.shape

for i in range(num_imgs):
    input_imgs[i] = path_to_input_image(input_img_paths[i])
    targets[i] = path_to_target_image(target_img_paths[i])

num_val_samples=1000
train_input_imgs = input_imgs[:-num_val_samples]
train_targets = targets[:-num_val_samples]
val_input_imgs = input_imgs[-num_val_samples:]
val_targets = targets[-num_val_samples:]
print(train_input_imgs.shape,train_targets.shape)
```

```python
num_classes = 3
inputs = keras.Input(shape=img_size + (3,))
x = layers.Rescaling(1./255.,name="my_rescaling")(inputs)
x = layers.Conv2D(64,3,strides=2,activation="relu",padding="same",name="my_first_conv2d")(x)
x = layers.Conv2D(64,3,activation="relu",padding="same",name="my_second_conv2d")(x)
x = layers.Conv2D(128,3,strides=2,activation="relu",padding="same",name="my_third_conv2d")(x)
x = layers.Conv2D(128,3,activation="relu",padding="same",name="my_fouth_conv2d")(x)
x = layers.Conv2D(256,3,strides=2,activation="relu",padding="same",name="my_fifth_conv2d")(x)
x = layers.Conv2D(256,3,activation="relu",padding="same",name="my_sixth_conv2d")(x)

x = layers.Conv2DTranspose(256,3,activation="relu",padding="same",name="my_first_conv2dtranpose")(x)
x = layers.Conv2DTranspose(256,3,strides=2,activation="relu",padding="same",name="my_second_conv2dtranpose")(x)
x = layers.Conv2DTranspose(128,3,activation="relu",padding="same",name="my_third_conv2dtranpose")(x)
x = layers.Conv2DTranspose(128,3,strides=2,activation="relu",padding="same",name="my_fouth_conv2dtranpose")(x)
x = layers.Conv2DTranspose(64,3,activation="relu",padding="same",name="my_fifth_conv2dtranpose")(x)
x = layers.Conv2DTranspose(64,3,strides=2,activation="relu",padding="same",name="my_sixth_conv2dtranpose")(x)

outputs = layers.Conv2D(num_classes,3,activation="softmax", padding="same")(x)
model = keras.Model(inputs,outputs)
model.summary()
keras.utils.plot_model(model)
```

```python
model.compile(optimizer="rmsprop",loss="sparse_categorical_crossentropy")
callbacks=[keras.callbacks.ModelCheckpoint("p20_segmentation_model.keras",save_best_only=True)]
history = model.fit(train_input_imgs,train_targets,
                    epochs=20,
                    callbacks=callbacks,
                    batch_size=64,
                    validation_data=(val_input_imgs,val_targets))
```

```python
import matplotlib.pyplot as plt
#print(history.__dict__)
epoch = history.epoch
loss = history.history["loss"]
val_loss = history.history["val_loss"]
loss = history.history["loss"]
print(epoch)
plt.figure()
plt.plot(epoch,loss,'r',label="loss")
plt.plot(epoch,val_loss,'g.',label="val_loss")
plt.legend()
plt.show()
```

```python
from tensorflow.keras.utils import array_to_img
model = keras.models.load_model("p20_segmentation_model.keras")
test_image = val_input_imgs[7]
plt.axis("off")
#plt.matshow(test_image)
#plt.imshow(test_image)
plt.imshow(array_to_img(test_image))

plt.figure()
prediction_batch = model.predict(np.expand_dims(test_image,0))
prediction = prediction_batch[0]
print(prediction_batch.shape,prediction[0,0,0],prediction[0,0,1],prediction[0,0,2])
def display_prediction(pred):
    mask = np.argmax(pred,axis=-1) # extract the index of the maximum of the softmax prediction, i.e. 0, 1, or 2
    mask *= 127                    # rescale from 0-2 back to 0-255 range
    #print(mask.shape)
    plt.axis("off")
    plt.imshow(mask)               # imshow can display a [0-255] array directly
display_prediction(prediction)    
```
